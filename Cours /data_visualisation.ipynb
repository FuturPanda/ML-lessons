## Considérations techniques : 

### Clareté : 

S'assurer que le jeu de donnée est complet et pertinent. 
1. Pas de valeurs nulles
2. Pas de doublons
3. Pas de valeurs manquantes. 


### Conseils : 

Utiliser les 4 chart les plus utilisées : 

1. Bar Chart 
2. Histogram 
3. Scatter plot
4. Line Chart

## RELATIONSHIPS

### Scatterplot : 

**Souvent utilisé pour montré la corrélation, la tendance, etc... entre deux variables**

Montre la relation entre deux variables (X,Y)
Variable indépendante X, variable dépendante Y 

### BubleChart : 

Montre la relation entre triois variables (X,Y,Bubble Size)

## Comparaisons 

### Bar Chart 
- Vertical Bar Chart : 
**Utilisée surtout pour la comparaison de catégorie.**
- Horizontal Bar Chart : 
**Utilisée surtout pour le ranking (ex : elections, sales performance stats, oscars etc...)**

Ex : 
Comparer les salaires : 

Y = Salaires 

X = Employés

### Line Chart 

**Souvent utilisée pour montre comment quelque chose change au cours du temps**

Comparer deux courbes (line), selon X et Y 
Ex : comparer les prix médians(Line1) et moyens(Line2) des prix(Y) de l'immobilier selon les années(X). 


## DISTRIBUTIONS 

### Histograms

Pour avoir les valeurs moyens et la déviation standard

### Box Plot 

Minimum et maximum values, median, et la majorité des données (50% Percentile). 

## COMPOSITIONS

### Pie Chart
**Bar charts are almost always better than pie chart**. 

Mais si on doit vraiment utiliser une pie chart : 

- Se limiter a 5 points de données
- Lister par ordre de grandeur depuis le plus large set orienté en haut a droite. 
- Ecrire a coté de chaque morceau le nombre et le nom de la donnée. 
- Utiliser la couleur

Composition d'une aire. 

### Stacked Bar Chart
**Utilisée pour représenter part to whole informations**

(Ex = bar chart pour représenter les telechargement des principales app de réseaux sociaux, mais stacked bar chart pour représenter les telechargement de chaque app selon les pays. )


Composition d'une aire selon une variable (ex composition des metiers au sein d'une entreprise selon les années)

### Stacked Area Chart
