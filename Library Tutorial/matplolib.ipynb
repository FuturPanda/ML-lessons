### Fonction Basique :

A noter : 
- Bien avoir des tableaux de mem dimensions

```python
x = np.linspace(0,2, 10)
y = x**2
# -- Pour une droite -- 
plt.plot(x,y)
#plt.show(x,y) -- A ecrire si pas dans jupyter nb 
```

```python
# -- Pour une nuage de points 
plt.scatter(x, y)
```

### Customisation : 

Pas trop de customisation : 
- c ='color'  -> Couleur de la courbe
- lw = nb  -> Epaisseur de la ligne (line width)
- ls = "--" -> style de la ligne. (line style)
- label = "name" -> Nom de la courbe. 


```python
plt.plot(x,y, label='tuning de graph',c='red', ls='--', )
```

## Le Cycle de vie d'une figure.
1. D'abord, on crée une figure. 
2. On indique la taille de la figure ; figsize =
3. On ajoute un titre, des label x et y et des label pour les courbes une legende, 

```python
plt.figure(figsize=[6,9])
plt.plot(x,y, label="yop")
plt.plot(x, y+7, label="Bloup")
plt.title("tuning de graph")
plt.xlabel("Modafucking X")
plt.ylabel("Whopping Y")
#plt.show()
#plt.save()
```
