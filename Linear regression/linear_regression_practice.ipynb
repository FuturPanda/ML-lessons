# Four : calculer la cost fonction. 

```python
def compute_cost(X, y, w, b):
    m = X.shape[0]
    cost = 0.
    for i in range(m):
      f_wb_i = np.dot(w, X[i]) + b 
      cost = cost + (f_wb_i - y[i])**2
    cost = cost/(2*m)
    print(f"Cost : {cost}")
    return cost

compute_cost(X_train, y_train, 0.3, 700)
```

# Five : Gradient Descent
On ne peut pas encore vraiment tester notre cost fonction, puisque nous n'avons encore aucunes valeurs de références pour w et b. 
On va donc entrainer le modèle pour essayer de trouver w et b optimaux. 

```python
def compute_gradient(X, y, w, b):
    m = X.shape[0]
    dj_dw = np.zeros(1)
    dj_db = 0
    for i in range(m):
        err = (np.dot(X[i], w) +b) - y[i]
        dj_dw = dj_dw + err * X[i]
        dj_db = dj_db + err
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    
    return dj_db, dj_dw

compute_gradient(X_train, y_train, 1, 1)
```

# Six : Gradient descent
On va computer encore et encore gradient pour apprendre les meilleures valeurs de w et b 

```python
def gradient_descent(X, y, w_in, b_in, cost_fonction, gradient_fonction, alpha, iters):
    J_history = []
    w = copy.deepcopy(w_in)
    b = b_in

    for i in range(iters) : 
        dj_db, dj_dw = gradient_fonction(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        print(f"w = {w}")
        print(f"b = {b}")

        if i < 10000:
            new_cost = cost_fonction(X, y, w, b)
            print(f'New cost = {new_cost}')
            J_history.append(new_cost)

        # if i% math.ceil(iters / 10) == 0:
        #     print(f"iteration {i:4d}: cost {J_history[-1]:8.2f}")
    return w, b, J_history
```
